```python
import os
import sys
import session_info
import geopandas as gpd
import numpy as np
import scipy
import pandas as pd
pd.set_option('future.no_silent_downcasting', True)
import matplotlib.pyplot as plt
import mpl_toolkits
import rtree
from shapely.geometry import Polygon

"""
Reads shape data from mid decade redistricting states
Only uses lower 48 states

"""

Shapepath = "Mid_decade_data/"
Dataoutpath ="Process data/"

Projection = "ESRI:102008"
#session_info.show()
print( "Libraries loaded")
```

```python
### Get CA data
File = 'AB604'
Shapedatafile = Shapepath + File + ".shp"
CA_districts = gpd.read_file( Shapedatafile)
CA_districts = CA_districts[['DISTRICT','geometry']]
CA_districts['STATEFP'] = '06'
CA_districts['SESSN'] = 'Mid-decade'
CA_districts['GEOID'] = CA_districts['STATEFP'] + CA_districts['DISTRICT']
CA_districts = CA_districts.reindex(columns=['GEOID','STATEFP','DISTRICT','SESSN','geometry'])
CA_districts.to_crs(Projection, inplace=True)
print('Done')
CA_districts.head()
```

```python
### Get TX data
File = 'PLANC2333'
Shapedatafile = Shapepath + File + ".shp"
TX_districts = gpd.read_file( Shapedatafile)
TX_districts.rename(columns={'District': 'DISTRICT'}, inplace=True)
TX_districts['DISTRICT'] = TX_districts['DISTRICT'].astype(str).str.zfill(2)
TX_districts['STATEFP'] = '48'
TX_districts['SESSN'] = 'Mid-decade'
TX_districts['GEOID'] = TX_districts['STATEFP'] + TX_districts['DISTRICT'] 
TX_districts = TX_districts.reindex(columns=['GEOID','STATEFP','DISTRICT','SESSN','geometry'])
TX_districts.to_crs(Projection, inplace=True)
print('Done')
TX_districts.head()
```

```python
### Get NC data
File = 'NC_Congress_2025'
Shapedatafile = Shapepath + File + ".shp"
NC_districts = gpd.read_file( Shapedatafile)
NC_districts = NC_districts[['District','geometry']]
NC_districts.rename(columns={'District': 'DISTRICT'}, inplace=True)
NC_districts['DISTRICT'] = NC_districts['DISTRICT'].astype(str).str.zfill(2)
NC_districts['STATEFP'] = '37'
NC_districts['SESSN'] = 'Mid-decade'
NC_districts['GEOID'] = NC_districts['STATEFP'] + NC_districts['DISTRICT'] 
NC_districts = NC_districts.reindex(columns=['GEOID','STATEFP','DISTRICT','SESSN','geometry'])
NC_districts.to_crs(Projection, inplace=True)
print('Done')
NC_districts.head()
```

```python
### Get MO data
File = 'HB1_Missouri_Congressional_Districts_2025'
Shapedatafile = Shapepath + File + ".shp"
MO_districts = gpd.read_file( Shapedatafile)
MO_districts = MO_districts[['District','geometry']]
MO_districts.rename(columns={'District': 'DISTRICT'}, inplace=True)
MO_districts['DISTRICT'] = MO_districts['DISTRICT'].astype(str).str.zfill(2)
MO_districts['STATEFP'] = '29'
MO_districts['SESSN'] = 'Mid-decade'
MO_districts['GEOID'] = MO_districts['STATEFP'] + MO_districts['DISTRICT'] 
MO_districts = MO_districts.reindex(columns=['GEOID','STATEFP','DISTRICT','SESSN','geometry'])
MO_districts.to_crs(Projection, inplace=True)
print('Done')
MO_districts.head()
```

```python
### Get VA data
File = 'VA cong VA Congressional Proposed - Amended (2026-02-13)'
Shapedatafile = Shapepath + File + ".shp"
VA_districts = gpd.read_file( Shapedatafile)
VA_districts.rename(columns={'district': 'DISTRICT'}, inplace=True)
VA_districts['DISTRICT'] = VA_districts['DISTRICT'].astype(str).str.zfill(2)
VA_districts['STATEFP'] = '51'
VA_districts['SESSN'] = 'Mid-decade'
VA_districts['GEOID'] = VA_districts['STATEFP'] + VA_districts['DISTRICT'] 
VA_districts = VA_districts.reindex(columns=['GEOID','STATEFP','DISTRICT','SESSN','geometry'])
VA_districts.to_crs(Projection, inplace=True)
print('Done')
VA_districts.head()
```

```python
### Get FL data
File = 'EOGPCRP2026'
Shapedatafile = Shapepath + File + ".shp"
FL_districts = gpd.read_file( Shapedatafile)
FL_districts['DISTRICT'] = FL_districts['DISTRICT'].astype(str).str.zfill(2)
FL_districts['STATEFP'] = '12'
FL_districts['SESSN'] = 'Mid-decade'
FL_districts['GEOID'] = FL_districts['STATEFP'] + FL_districts['DISTRICT'] 
FL_districts = FL_districts.reindex(columns=['GEOID','STATEFP','DISTRICT','SESSN','geometry'])
FL_districts.to_crs(Projection, inplace=True)
print('Done')
FL_districts.head()
```

```python
#### Combine and save file for further processing
Dfs = [CA_districts,TX_districts,NC_districts,MO_districts,VA_districts, FL_districts]
Mid_decade_districts = pd.concat( Dfs, ignore_index=True)
Shapeoutfile = Dataoutpath + 'Mid_decade'  + ".shp"
Mid_decade_districts.to_file( Shapeoutfile )
print( 'Done writing ', len( Mid_decade_districts ), 'rows to ', Shapeoutfile)
```
